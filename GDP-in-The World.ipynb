```python
def classification(nation):
    xx = pd.Series(gdp_data.loc[nation])
    xx = xx.replace(to_replace = '..', value=0)
    xx = xx.astype('float32')
    return xx
```

```python
GDP_Uni_Eropa = []

for i in EU:
    a = classification(i)
    if len(GDP_Uni_Eropa) == 0:
        GDP_Uni_Eropa = a
    else:
        GDP_Uni_Eropa = GDP_Uni_Eropa + a
```

```python
GDP_Amerika_Utara = []

for i in North_America_nation:
    a = classification(i)
    if len(GDP_Amerika_Utara) == 0:
        GDP_Amerika_Utara = a
    else:
        GDP_Amerika_Utara = GDP_Amerika_Utara + a
```

```python
GDP_Amerika_Selatan = []

for i in South_America_nation:
    a = classification(i)
    if len(GDP_Amerika_Selatan) == 0:
        GDP_Amerika_Selatan = a
    else:
        GDP_Amerika_Selatan = GDP_Amerika_Selatan + a
```

```python
GDP_Amerika_Tengah = []

for i in Central_America_nation:
    a = classification(i)
    if len(GDP_Amerika_Tengah) == 0:
        GDP_Amerika_Tengah = a
    else:
        GDP_Amerika_Tengah = GDP_Amerika_Tengah + a
```

```python
GDP_Asia = []

for i in country_of_Asia:
    if i in gdp_data.index:
        a = classification(i)
        if len(GDP_Asia) == 0:
            GDP_Asia = a
        else:
            GDP_Asia = GDP_Asia + a
    else:
        continue
```

```python
GDP_G20_Asia = []

for i in G20_Asia_nation:
    if i in gdp_data.index:
        a = classification(i)
        if len(GDP_G20_Asia) == 0:
            GDP_G20_Asia = a
        else:
            GDP_G20_Asia = GDP_G20_Asia + a
            
GDP_G20_Asia
```
